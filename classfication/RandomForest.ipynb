```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
# df=pd.read_csv('test_Y3wMUE5_7gLdaTN.csv')
```

```python
df.isnull().sum()
df.info()
```

```python
char_col=df.select_dtypes(include='object').columns
```

```python
q1,q3=df['ApplicantIncome'].quantile(0.25),df['ApplicantIncome'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-iqr*1.5
upper_bound=q3+iqr*1.5
df['ApplicantIncome']=df['ApplicantIncome'].apply(lambda x:
    upper_bound if x > upper_bound else( lower_bound if x < lower_bound else x) )
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
ll=LabelEncoder()
encoding=['Loan_Status','Property_Area','Education','Married','Self_Employed','Gender','Dependents']
for col in encoding:
    df[col]=ll.fit_transform(df[[col]])
```

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
df['Gender']=imputer.fit_transform(df[['Gender']]).ravel()
df['Married']=imputer.fit_transform(df[['Married']]).ravel()
df['Self_Employed']=imputer.fit_transform(df[['Self_Employed']]).ravel()
df['Dependents']=imputer.fit_transform(df[['Dependents']]).ravel()
imputer=SimpleImputer(strategy='median')
df['LoanAmount']=imputer.fit_transform(df[['LoanAmount']]).ravel()
# df['Education']=imputer.fit_transform(df[['Education']]).ravel()
df['Loan_Amount_Term']=imputer.fit_transform(df[['Loan_Amount_Term']]).ravel()
df['Credit_History']=imputer.fit_transform(df[['Credit_History']]).ravel()
```

```python
df_new=df.drop(['Loan_ID'],axis=True)
```

```python
df_new.corr()['Loan_Status'].sort_values(ascending=False)
x,y=df_new.drop('Loan_Status',axis=True),df_new['Loan_Status']
```

```python
df_new.head(5)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn.metrics import classification_report,accuracy_score
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
# from sklearn.preprocessing import
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model =RandomForestClassifier()
x_train=ss.fit_transform(x_train)
model.fit(x_train,y_train)
x_test=ss.transform(x_test)
y_pred=model.predict(x_test)
print(f"{accuracy_score(y_test,y_pred)}\n{classification_report(y_test,y_pred)}")
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
param_dist = {
    'n_estimators': [50, 100, 200, 300, 400],
    'max_depth': [None, 5, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
# random_grid=RandomizedSearchCV(RandomForestClassifier(random_state=42),param_distributions=param_dist,cv=5,random_state=42)
random_grid=RandomizedSearchCV( estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_dist,
    n_iter=20,
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1)
random_grid.fit(x_train,y_train)

print("Best Parameters:", random_grid.best_params_)

# Best estimator
best_model = random_grid.best_estimator_

# Predict (on training data for now)
y_pred = best_model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
